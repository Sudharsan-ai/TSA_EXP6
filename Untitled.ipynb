```python
# Developed by: Your Name
# Register Number: Your Reg Number

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.seasonal import seasonal_decompose

# ============================
# STEP 1: File Upload / Data Loading
# ============================

# Load dataset (must have Date and Amount columns)
data = pd.read_csv("Amazon Sale Report.csv", parse_dates=["Date"])

# If only Date is present, create dummy Amount values
if "Amount" not in data.columns:
    print("⚠ No 'Amount' column found. Creating dummy values...")
    data["Amount"] = np.random.randint(100, 500, size=len(data))

# Set Date as index
data.set_index("Date", inplace=True)

# Convert to weekly data (resample by week, sum sales)
data_weekly = data["Amount"].resample("W").sum()

# ============================
# STEP 2: Scaling the data
# ============================
scaler = MinMaxScaler()
scaled_data = pd.Series(
    scaler.fit_transform(data_weekly.values.reshape(-1, 1)).flatten(),
    index=data_weekly.index
)

plt.figure(figsize=(10,5))
scaled_data.plot(color="blue")
plt.title("Scaled Weekly Sales Data (Date vs Mean Amount)", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Mean Amount", fontsize=12)   # Y-axis = Mean
plt.grid(True)
plt.show()

# ============================
# STEP 3: Seasonal Decomposition (only if enough weeks)
# ============================
if len(data_weekly) >= 8:  # Need at least 8 weeks
    decomposition = seasonal_decompose(data_weekly, model="additive", period=4)
    fig = decomposition.plot()
    fig.set_size_inches(10, 8)
    plt.suptitle("Seasonal Decomposition (Weekly Additive Model)", fontsize=16)

    for ax, label in zip(fig.axes, ["Observed", "Trend", "Seasonal", "Residual"]):
        ax.set_ylabel(f"{label} (Mean)", fontsize=12)   # Add Mean to labels
        ax.set_xlabel("Date", fontsize=12)

    plt.show()
else:
    print("⚠ Not enough weekly data for seasonal decomposition (need at least 8 weeks). Skipping...")

# ============================
# STEP 4: Train-Test Split
# ============================
scaled_data = scaled_data + 1  # Ensure positivity
train_data = scaled_data[:int(len(scaled_data) * 0.8)]
test_data = scaled_data[int(len(scaled_data) * 0.8):]

# ============================
# STEP 5: Model Training & Forecasting
# ============================
if len(train_data) >= 4:  # At least one seasonal cycle (4 weeks)
    model_add = ExponentialSmoothing(
        train_data, trend='add', seasonal='add', seasonal_periods=4
    ).fit()

    test_predictions_add = model_add.forecast(steps=len(test_data))

    plt.figure(figsize=(10,5))
    train_data.plot(label="Train Data", color="blue")
    test_predictions_add.plot(label="Test Predictions", color="green")
    test_data.plot(label="Test Data", color="red")
    plt.legend()
    plt.title("Train vs Test Predictions (Weekly Sales)", fontsize=14)
    plt.xlabel("Date", fontsize=12)
    plt.ylabel("Mean Amount", fontsize=12)   # Y-axis = Mean
    plt.grid(True)
    plt.show()

    # ============================
    # STEP 6: Model Evaluation
    # ============================
    rmse = np.sqrt(mean_squared_error(test_data, test_predictions_add))
    print("RMSE:", rmse)
    print("Reference (Std Dev, Mean):", np.sqrt(scaled_data.var()), scaled_data.mean())
else:
    print("⚠ Not enough weekly data to train Holt-Winters model (need at least 4 weeks). Skipping model training.")

# ============================
# STEP 7: Final Model for Full Dataset
# ============================
if len(data_weekly) >= 4:
    final_model = ExponentialSmoothing(
        data_weekly, trend='add', seasonal='mul', seasonal_periods=4
    ).fit()

    final_predictions = final_model.forecast(steps=8)  # Forecast next 8 weeks (~2 months)

    plt.figure(figsize=(10,5))
    data_weekly.plot(label="Original Weekly Sales", color="blue")
    final_predictions.plot(label="Final Predictions", color="orange")
    plt.legend()
    plt.title("Final Forecast (Weekly Sales - Date vs Mean Amount)", fontsize=14)
    plt.xlabel("Date", fontsize=12)
    plt.ylabel("Mean Amount", fontsize=12)   # Y-axis = Mean
    plt.grid(True)
    plt.show()
else:
    print("⚠ Not enough weekly data to build final forecast (need at least 4 weeks).")
```
